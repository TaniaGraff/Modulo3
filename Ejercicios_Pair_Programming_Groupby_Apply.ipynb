```python
#Como con apply me da error y me convierte todos los valores a nan, lo hago manualmente, columna por columna.
#Columna density.
df_world['density'] = df_world['density'].astype(str).str.replace(',', '').astype(float)

#Columna landarea.
try:
    df_world['landarea'] = df_world['landarea'].astype(str).str.replace(',', '').astype(float)
except ValueError:
    df_world['landarea'] = np.nan

df_world['landarea']

#Columna armedforcessize.
try:
    df_world['armedforcessize'] = df_world['armedforcessize'].astype(str).str.replace(',', '').astype(float)
except ValueError:
    df_world['armedforcessize'] = np.nan

df_world['armedforcessize']

#Columna co2emissions.
try:
    df_world['co2emissions'] = df_world['co2emissions'].astype(str).str.replace(',', '').astype(float)
except ValueError:
    df_world['co2emissions'] = np.nan

df_world['co2emissions']

#Columna population.
try:
    df_world['population'] = df_world['population'].astype(str).str.replace(',', '').astype(float)
except ValueError:
    df_world['population'] = np.nan

df_world['population']

#Columna urbanpopulation.
try:
    df_world['urbanpopulation'] = df_world['urbanpopulation'].astype(str).str.replace(',', '').astype(float)
except ValueError:
    df_world['urbanpopulation'] = np.nan

df_world['urbanpopulation']

#Columna gdp.
try:
    df_world['gdp'] = df_world['gdp'].astype(str).str.replace(',', '').astype(float)
except ValueError:
    df_world['gdp'] = np.nan

df_world['gdp']

#Columna cpi. 
try:
    df_world['cpi'] = df_world['cpi'].astype(str).str.replace(',', '').astype(float)
except ValueError:
    df_world['cpi'] = np.nan

#Imprimo columnas para comprobar los valores
df_world.dtypes
```

```python
#Sigo con las dos últimas columnas que faltan. Cambio y limpio la columna latitude.
try:
    df_world['latitude'] = df_world['latitude'].astype(str).str.split(')').str[0].str.replace("'", "").replace(',', '').astype(float)
except ValueError:
    df_world['latitude'] = np.nan

df_world['latitude']
```

```python
# Cambio y limpio la columna longitude.
try:
    df_world['longitude'] = df_world['longitude'].astype(str).str.replace("'", "").str.replace("(", "").str.replace(")", "").str.replace(",", ".").astype(float)
except ValueError:
    df_world['longitude'] = np.nan

df_world['longitude']
```

5. ¿Cuál es la tasa de impuestos total máxima por densidad de población por continente? Para solucionar este ejercicio puedes:

8. ¿Cuál es la tasa de maternidad promedio por tasa de impuestos total? Para resolver el ejercicio deberás:

- Primero, calcula la tasa de maternidad promedio para cada continente. Puedes usar la columna "maternal_mortality_ratio" que representa la tasa de mortalidad materna.

- Calcula la tasa de impuestos total promedio para cada continente. Puedes usar la columna "total_tax_rate" que representa la tasa de impuestos total en cada continente.

- Luego, divide la tasa de maternidad promedio por la tasa de impuestos total promedio para obtener la relación entre ambas tasas.

- Encuentra el país con la tasa de maternidad más alta por punto porcentual de tasa de impuestos total.

```python
#Calculo la tasa de maternidad promedio para cada continente.
media_maternidad_por_continente = df_world_continentes.groupby('continent')['maternalmortalityratio'].mean()

#Calculo la tasa de impuestos total promedio para cada continente.
promedio_impuestos_por_continente = df_world_continentes.groupby('continent')['totaltaxrate'].mean()

#Calculo la relación entre la tasa de maternidad promedio y la tasa de impuestos total promedio.
tasa_maternidad_por_impuestos = media_maternidad_por_continente / promedio_impuestos_por_continente

tasa_maternidad_por_impuestos.reset_index(name='maternityratebytaxes')
```

```python
# Ahora, para encontrar el país con la tasa de maternidad más alta por punto porcentual de tasa de impuestos total
df_world_continentes['tasa_maternidad_por_impuestos'] = df_world_continentes['maternalmortalityratio'] / df_world_continentes['totaltaxrate']

# Encontrar el país con la tasa de maternidad más alta por punto porcentual de tasa de impuestos total
pais_con_mayor_tasa = df_world_continentes.loc[df_world_continentes['tasa_maternidad_por_impuestos'].idxmax()]

pais_con_mayor_tasa[['country', 'tasa_maternidad_por_impuestos']].reset_index(name='countrywithhighestmaternityrate')

```
