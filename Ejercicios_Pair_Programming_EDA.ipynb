3. Ejercicios con loc e iloc
- Selecciona todas las columnas de las filas 2 a 4 (inclusive) utilizando loc.

- Utilizando loc, filtra el conjunto de datos para mostrar solo las filas donde la densidad (columna "Birth Rate") es mayor a 30.

- Utilizando loc selecciona las columnas "Country" y "Capital/Major City" de las filas donde el porcentaje de tierra agrícola (columna "Fertility Rate") es superior al 3.

- Selecciona todas las filas de las columnas 0 a 4 (excluyendo la columna 4) utilizando iloc.

- Encuentra el valor en la fila 2 y columna 7 utilizando iloc.

- Selecciona todas las filas para las columnas 0, 2, y 5 utilizando iloc.

```python
#Selecciono todas las columnas de las filas 2 a 4 (inclusive) utilizando loc.
df_world.loc['Albania':'Angola']
#Entiendo que la opción anterior no sale porque son los valores de la columna Country, así que tendría que entrar en la columna Country para obtenerlos, pero si entro, no sé como usar el star, stop, step.
df_world.loc[(df_world.Country == 'Albania') | (df_world.Country == 'Algeria') | (df_world.Country == 'Andorra') ]
```

```python
#Utilizando loc, filtro el conjunto de datos para mostrar solo las filas donde la densidad (columna "Birth Rate") es mayor a 30.
densidad= df_world.loc[df_world['Birth Rate'] > 30, ['Country','Birth Rate']]
densidad.head()
```

```python
#Utilizando loc selecciona las columnas "Country" y "Capital/Major City" de las filas donde el porcentaje de tierra agrícola (columna "Fertility Rate") es superior al 3.
fertilidad = df_world.loc[df_world['Fertility Rate']>3, ['Country', 'Capital/Major City', 'Fertility Rate']]
fertilidad.head()
```

```python
#Selecciona todas las filas de las columnas 0 a 4 (excluyendo la columna 4) utilizando iloc.
df_world.iloc[:,:4]
```

```python
#Encuentra el valor en la fila 2 y columna 7 utilizando iloc.
df_world.iloc[1,6]
```

```python
#Selecciona todas las filas para las columnas 0, 2, y 5 utilizando iloc. Como habla de la columna 0 entiendo que se refiere al índice de las columnas.
df_world.iloc[:, [0,2,5]]
```

4. BONUS DE LOC Y ILOC
- Encuentra el valor de "Gasoline Price" para el país cuya abreviatura es "AL" utilizando loc.

- Selecciona todas las columnas para el país cuya capital es "Tirana" utilizando loc.

- Filtra el conjunto de datos para mostrar solo las filas donde el índice de CPI (columna "CPI") sea igual a 128.85.

- Encuentra la densidad (columna "Density\n(P/Km2)") para el país cuya abreviatura es "AD" utilizando loc.

- Selecciona las columnas "Country" y "Co2-Emissions".

- Filtra el conjunto de datos para mostrar solo las filas donde el índice de CPI cambió un 2% (columna "CPI Change (%)").

- Encuentra la tasa de natalidad (columna "Birth Rate") para el país con la abreviatura "AF" utilizando loc.

- Selecciona todas las columnas de las filas 3 a 5 (excluyendo la fila 5) utilizando iloc.

- Selecciona las columnas 1, 3 y 6 para todas las filas utilizando iloc.

```python
#Busco el valor de "Gasoline Price" para el país cuya abreviatura es "AL" utilizando loc.
df_world.loc[df_world['Abbreviation']=='AL', ['Country', 'Abbreviation', 'Gasoline Price']]
```

```python
#Selecciono todas las columnas para el país cuya capital es "Tirana" utilizando loc.
df_world.loc[df_world['Capital/Major City']== 'Tirana', :]
```

```python
#Muestro únicamente las filas donde el índice de CPI (columna "CPI") sea igual a 128.85. Uso comillas para seleccionar el valor supuestamente numérico porque en realidad es un object o string.
df_world.loc[(df_world.CPI == '128.85'), ['Country', 'CPI']]
```

```python
#Encuentra la densidad (columna "Density\n(P/Km2)") para el país cuya abreviatura es "AD" utilizando loc.
df_world.loc[df_world['Abbreviation']== 'AD', ['Country', 'Abbreviation', 'Density\n(P/Km2)']]
```

```python
#Selecciona las columnas "Country" y "Co2-Emissions".
df_world[['Country', 'Co2-Emissions']]
```

```python
#Filtra el conjunto de datos para mostrar solo las filas donde el índice de CPI cambió un 2% (columna "CPI Change (%)").
#Imprimo la columna para ver como están inscritos los valores, de este modo observo que son tipo object, aunque estén escritos como float. De ahí que al 2, le añada dos ceros de decimales. 
df_world.loc[df_world['CPI Change (%)']== '2.00%', ['Country', 'CPI Change (%)']]
```

```python
#Busco la tasa de natalidad (columna "Birth Rate") para el país con la abreviatura "AF" utilizando loc.
df_world.loc[(df_world.Abbreviation== 'AF'), ['Country', 'Abbreviation', 'Birth Rate']]
```

```python
#Selecciono todas las columnas de las filas 3 a 5 (excluyendo la fila 5) utilizando iloc.
df_world.iloc[2:5, :]
```

```python
#Selecciono las columnas 1, 3 y 6 para todas las filas utilizando iloc. Uso el índice 6 porque no eliminé del DF la columna de Armed Forces Size.
df_world.iloc[:, [0,2,6]]
```
